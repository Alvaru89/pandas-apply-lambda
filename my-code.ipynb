```python
df["Bin"]=df.apply(lambda x: vote_bin(x["Votes"]) , axis=1 )
```

```python
def new_rank(e,f):
    if "Thriller" in e:
        f+=1
    if "Comedy" in e:
        f-=1
    return f

df["New Rating"]=df.apply(lambda x: new_rank(x["Genre"], x["Rating"]) , axis=1 )
df.sort_values(by=["New Rating"],ascending=False)
```

Challenge 4

```python
def chall_4(title, votes):
    ascii_sum=0
    for letter in title:
        ascii_sum+=ord(letter)
    if ascii_sum%votes!=0:
        return False
    return is_prime(ascii_sum/votes)
    
```

```python
def is_prime(n):
    check=0
    i=n-1
    while i>1:
        if n%i==0:
            check=1
        i-=1
    return check==0

print(is_prime(5))
print(is_prime(2))
print(is_prime(1))
print(is_prime(6))
```

```python
df["ASCII/Votes=prime"]=df.apply(lambda x: chall_4(x["Title"], x["Votes"]) , axis=1 )
```

Challenge 6

```python
from hashlib import sha256

def freaky_bonus(a,b,c):
    hashedWord = sha256(a.encode('UTF-8')).hexdigest()
    sum=0
    for i in hashedWord:
        try:
            int(i)
            sum+=i
        except:
            pass
    return sum>b and sum<(c*1000000)

df["Hidden Pattern"]=df.apply(lambda x: freaky_bonus(x["Description"], x["Votes"],x["Revenue (Millions)"]) , axis=1 )
df["Hidden Pattern"].value_counts()
```
